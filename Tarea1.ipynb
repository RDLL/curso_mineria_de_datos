Creamos la variable df y le pasamos el dataset 

```python
df= pd.read_csv('https://pkgstore.datahub.io/machine-learning/cervical-cancer/cervical-cancer_csv/data/c43a91f09832cd9db8db84df08c942dc/cervical-cancer_csv.csv')
```

Mostramos las primeras 10 columnas

```python
df.head(n=10)
```

Reemplazamos los valores numericos `0,1,2,3 y 4` de la columna `STDs (number)`

con `"muy bajo", "bajo", "medio", "alto" y "muy alto"`

```python
df["STDs (number)"]=df["STDs (number)"].replace([0,1,2,3,4],["muy bajo","bajo","medio","alto","muy alto"])
```

```python
df["STDs (number)"]
```

Obtener porcentaje de faltantes

```python
def faltantes(data):
  ft = data.isnull().sum(min_count=1)
  ft = ft[ft != 0].to_dict()
  return ft
```

```python
vfal = faltantes(df)
print(vfal)
porcentajes = vfal
for clave in vfal:
  porcentajes[clave] = (vfal[clave]/df.shape[0]) * 100
print(porcentajes)
```

Eliminacion de registros nulos en 'Smokes'

```python
df = df.dropna(subset=["Smokes"])
compf = faltantes(df)
print(compf)
```

Eliminacion de columnas con mas de 30% de valores nulos

```python
for clave in porcentajes:
  if porcentajes[clave] > 30:
    del vfal[clave]
    df = df.drop([clave],axis = 1)
print(df.dtypes)
```

Obtener la media aritmetica y remplazar los valores nulo con la media de las columnas numericas

```python
means = df.mean(numeric_only=True).to_dict()

df = df.fillna(means)

print(faltantes(df))
```

Reemplazar los valores de la columna categorica por la moda de dicha columna

```python
moda = str(df["STDs (number)"].mode())
df["STDs (number)"]=df["STDs (number)"].replace(["muy bajo","bajo","medio","alto","muy alto"],moda)
print(faltantes(df))
#La categorica la sigue marcando como si tuviera valores nulos
df.head(n=15)
```
